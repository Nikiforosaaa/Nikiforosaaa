```python
import numpy as np
import matplotlib.pyplot as plt

Tstar = np.linspace(0.01, 4, 500)
U_red = 0.5 / np.tanh(1/(2*Tstar))

plt.figure(figsize=(8, 6))
plt.plot(Tstar, U_red, color='black', lw=2)
plt.xlabel(r'Reduced temperature $T^* = k_B T / (\hbar \omega_0)$')
plt.ylabel(r'$U / (\hbar \omega_0)$')
plt.title('Internal Energy versus Reduced Temperature', fontsize=13)
plt.xlim(0, 4)
plt.ylim(0, 4)
plt.grid(True, linestyle=':', alpha=0.5)

plt.figtext(0.5, 0.18,
    "Parameters:\n"
    "ℏω₀ : quantum of energy    kB : Boltzmann constant\n"
    "T* = kB T / (ℏω₀) : reduced temperature",
    wrap=True, ha='center', va='bottom', fontsize=10, bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'))
plt.figtext(0.5, 0.13,
    r'$\frac{U}{\hbar \omega_0} = \frac{1}{2} \coth\left( \frac{1}{2T^*} \right)$',
    ha='center', va='bottom', fontsize=11)

plt.figtext(0.5, -0.02, 'NurAnggitaSari-2415011061', ha='center', va='bottom', fontsize=9, color='black')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

Tstar = np.linspace(0.01, 4, 500)
CV_red = (1/(4 * Tstar**2)) * (1 / np.sinh(1/(2*Tstar)))**2

plt.figure(figsize=(8, 6))
plt.plot(Tstar, CV_red, color='black', lw=2)
plt.xlabel(r'Reduced temperature $T^* = k_B T / (\hbar \omega_0)$')
plt.ylabel(r'$C_V / k_B$')
plt.title('Heat Capacity versus Reduced Temperature', fontsize=13)
plt.xlim(0, 4)
plt.ylim(0, 1.1)
plt.grid(True, linestyle=':', alpha=0.5)

plt.figtext(0.5, 0.18,
    "Parameters:\n"
    "kB : Boltzmann constant    ℏω₀ : quantum of energy\n"
    "T* = kB T / (ℏω₀) : reduced temperature",
    wrap=True, ha='center', va='bottom', fontsize=10, bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'))

plt.figtext(0.5, 0.13,
    r'$\frac{C_V}{k_B} = \frac{1}{4(T^*)^2} \mathrm{csch}^2\left( \frac{1}{2T^*} \right)$',
    ha='center', va='bottom', fontsize=11)

plt.figtext(0.5, -0.02, 'NurAnggitaSari-2415011061', ha='center', va='bottom', fontsize=9, color='black')

plt.show()
```
