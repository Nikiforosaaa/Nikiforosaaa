```python

# Problem 1 (Trapezoidal Shape)
# by Nur Anggita Sari

import numpy as np
import matplotlib.pyplot as plt

a_val = 0.25  # [0, 0.5]

def d(a):
    return np.sqrt((1 + 4 * a) / 4)
d_val = d(a_val)

def u_a(x):
    if x < a_val:
        return d_val - (d_val / a_val) * x
    elif x <= 1 - a_val:
        return 0
    else:
        return (d_val / a_val) * (x - 1) + d_val

u_vec = np.vectorize(u_a)

x = np.linspace(0, 1, 1000)
u = u_vec(x)

plt.figure(figsize=(6, 3.5))
plt.plot(x, u, label=fr'$u_a(x)$ with $a = {a_val}$', color='darkblue', linewidth=2.0)
plt.title(r"Minimize $J(u) = \int_0^1 u(x)\,dx$ under $L(u_a) = 2$ (Trapezoid shape)", fontsize=11)
plt.xlabel("$x$", fontsize=11)
plt.ylabel("$u_a(x)$", fontsize=11)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Problem 2(a) for [0,1]
# by Nur Anggita Sari

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 1000)
A = 1 / (np.exp(1) - np.exp(-1))
u = A * (np.exp(x) - np.exp(-x))

plt.figure(figsize=(6, 3.5))
plt.plot(x, u, label = r'$u(x) = \frac{e^{x}-e^{-x}}{e-e^{-1}}$', color='darkblue', linewidth=2.0)
plt.title("Minimize $J(u)=\int_{0}^{1} (u(x)^2 + (u'(x)^2)\,dx$ over $u(x)$ (EL)", fontsize=11)
plt.xlabel("($x$)", fontsize=11)
plt.ylabel("$u(x)$", fontsize=11)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Problem 2(b) for [-1,1]
# by Nur Anggita Sari

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 1000)
A = 1 / (np.exp(1) - np.exp(-1))
u = A * 2 * np.cosh(x)

plt.figure(figsize=(6, 3.5))
plt.plot(x, u, label = r'$u(x) = \frac{2 \cosh x}{e+e^{-1}}$', color='darkblue', linewidth=2.0)
plt.title("Minimize $J(u)=\int_{-1}^{1} (u(x)^2 + (u'(x))^2) \,dx$ over $u(x)$ (EL)", fontsize=11)
plt.xlabel("($x$)", fontsize=11)
plt.ylabel("$u(x)$", fontsize=11)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend()
plt.tight_layout()
plt.show()
```
