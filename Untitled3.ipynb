```python

import numpy as np
from scipy.fftpack import fftn, ifftn, fftshift, ifftshift
import matplotlib.pyplot as plt
```

```python
#object
dy, dx = 1e-6, 1e-6
wl = 532e-9
ny, nx = 128, 128

u0 = np.zeros((ny,nx), dtype=np.complex128)
w=10
u0[ny//2-w:ny//2+w, nx//2-w:nx//2+w] = 1
```

```python
plt.imshow(abs(u0))
```

```python
def fresnel_diffaction_conv1(u, dx, dy, w1, z):
  #Impulse response
  z = 1e-3
  ny, nx = u0.shape
  x = np.arange((-nx // 2 ), (nx+1) // 2) * dx
  y = np.arange((-ny // 2), (ny+1) // 2) * dy
  yy, xx = np.meshgrid(y, x, indexing='ij')

  print(yy.shape)
#print(y.shape)

  k = 2 * np.pi/wl
  h = np.exp(1j*k*z) / (1j*z) * np.exp(1j*k/(2*z) * (yy**2 + xx**2))

#zero padding
  u_zpad = np.pad(u0, ((ny//2, ny//2), (nx//2, nx//2)), mode='constant')
  h_zpad = np.pad(h, ((ny//2, ny//2), (nx//2, nx//2)), mode='constant')

#fftconvolve
  U = fftn(u_zpad)
  H = fftn(fftshift(h_zpad))
  u_out_zpad = ifftn(U*H)

#extraction
  c = (ny, nx)
  w = (ny //2, nx // 2)
  u_out = u_out_zpad[w[0]:w[0]+c[0], w[1]:w[1]+c[1]]

  print(u0.shape, u_zpad.shape)

#print(h.shape

  return u_out
```

```python
u_out = fresnel_diffaction_conv1(u0, dx, dy, wl, z=1e-3)
plt.imshow(np.abs(u_out))
```
